```python
from sklearn.tree import DecisionTreeClassifier

# feature =wheather ,wind 
#  wheater: 0 =sunny  ,1 rainy day
# wind : 0 =weak .1 strong
x=[
    [0,0], # wheater,wind
    [0,1],
    [1,0],
    [1,1]

]
# 1=play 0=not play
y=[1,0,1,0]

model=DecisionTreeClassifier()
model.fit(x,y)

wheater=int(input("wheater(0=suuny:1=rainy):"))
wind=int(input("wind(0=weak:1=strong):"))

result=model.predict([[wheater,wind]])[0]
if result==1:
    print("you can play")
else:
    print("you can't play")
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data={
    'color':['red','green','blue','green'],
    'size':['s','m','l','s'],
    'price':[100,200,300,400]
}
df=pd.DataFrame(data)
encoder=OneHotEncoder(sparse_output=False)

encoded=encoder.fit_transform(df[['color','size']])

encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(['color','size']))
final_df=pd.concat([df,encoded_df],axis=1)
final_df
```

```python
import pandas as pd
data={
    "age":[20,22,25,30,35],
    "salary":[30000,35000,40000,50000,60000]
}
df=pd.DataFrame(data)
df

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
scaled_df=pd.DataFrame(scaled_data,columns=df.columns)
scaled_df
```
