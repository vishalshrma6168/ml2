```python
from sklearn.tree import DecisionTreeClassifier

# feature =wheather ,wind 
#  wheater: 0 =sunny  ,1 rainy day
# wind : 0 =weak .1 strong
x=[
    [0,0], # wheater,wind
    [0,1],
    [1,0],
    [1,1]

]
# 1=play 0=not play
y=[1,0,1,0]

model=DecisionTreeClassifier()
model.fit(x,y)

wheater=int(input("wheater(0=suuny:1=rainy):"))
wind=int(input("wind(0=weak:1=strong):"))

result=model.predict([[wheater,wind]])[0]
if result==1:
    print("you can play")
else:
    print("you can't play")
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data={
    'color':['red','green','blue','green'],
    'size':['s','m','l','s'],
    'price':[100,200,300,400]
}
df=pd.DataFrame(data)
encoder=OneHotEncoder(sparse_output=False)

encoded=encoder.fit_transform(df[['color','size']])

encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(['color','size']))
final_df=pd.concat([df,encoded_df],axis=1)
final_df
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np
# age ,salary
x=np.array([
    [22,20000],
    [25,22000],
    [47,50000],
    [52,52000],
    [46,49500],
    [56,60000],
    [23,21000],
    [24,23000]
])
# y=0 low income,1=high income
y=np.array([0,0,1,1,1,1,0,0])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

cm=confusion_matrix(y_test,y_pred)
cm


disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['low','high'])
disp.plot()
plt.show()
```
