```python
import pandas as pd

data={
    "name":["a","b","c","d"],
    "age":[12,None,44,None],
    "salary":[14000,None,23000,66000]
}
df=pd.DataFrame(data)
df["age"].fillna(df["age"].mean(),inplace=True)
df.dropna()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data={
    "color":["red","green","pink","yellow","red","pink"],
    "size":["s","m","l","m","s","l"],
    "price":[100,150,130,120,160,130]
}
df=pd.DataFrame(data)
le=LabelEncoder()
df["color_encode"]=le.fit_transform(df["color"])
df["size_encode"]=le.fit_transform(df["size"])
df
```

```python
import pandas as pd

data={
    "color":["red","green","pink","yellow","red","pink"],
    "size":["s","m","l","m","s","l"],
    "price":[100,150,130,120,160,130]
}

df=pd.DataFrame(data)
df
# one hot encoding
df_encoded=pd.get_dummies(df,columns=['color','size'],drop_first=False)

df_encoded
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data={
    "color":["red","green","pink","yellow","red","pink"],
    "size":["s","m","l","m","s","l"],
    "price":[100,150,130,120,160,130]
}

df=pd.DataFrame(data)

encoder=OneHotEncoder(sparse_output=False)

encoded=encoder.fit_transform(df[['color','size']])

encode_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(['color','size']))

final_df=pd.concat([df,encode_df],axis=1)
final_df
```

```python
import pandas as pd
data={
    'age':[20,22,25,30,35],
    'salary':[30000,35000,40000,50000,60000]
}
df=pd.DataFrame(data)
df



from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
scaled_df=pd.DataFrame(scaled_data,columns=df.columns)
scaled_df
```
